#### **Univariate Stats**

```python
def unistats(df):
    output_df = pd.DataFrame(columns=["attribute_type", "count", "missing", "missing_percent", "unique", "dtype", "mean", "mode",
                                      "min", "25_quartile", "median", "75_quartile",
                                      "max", "std", "skew", "kurt"])
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            output_df.loc[col] = ["Numeric", df[col].count(), df[col].isnull().sum(), df[col].isnull().sum()/df.shape[0], df[col].nunique(), df[col].dtype, df[col].mean(), df[col].mode().values[0],
                                 df[col].min(), df[col].quantile(0.25), df[col].median(), df[col].quantile(0.75),
                                  df[col].max(), df[col].std(), skew(df[col], bias=False), df[col].kurt()]
        else:
            output_df.loc[col] = ["Categorical", df[col].count(), df[col].isnull().sum(), df[col].isnull().sum()/df.shape[0], df[col].nunique(), df[col].dtype, "", "", "", "", "", "", "", "", "", ""]
        
    return output_df.sort_values(["attribute_type", "skew", "unique"], ascending=False)
            
```

```python
stats_df = unistats(df)
stats_df.head(100)
```

#### **Bivariate Stats**

##### **Numeric:Numeric - Correlation**

```python
def bivariate_stat(df, label):
    corr_df = pd.DataFrame(columns=["r", "p"])
    for col in df.columns:
        if col == label:
            continue
        if df[col].isnull().sum() == 0:
            if pd.api.types.is_numeric_dtype(df[col]):
                r, p = pearsonr(df[label], df[col])
                corr_df.loc[col] = [round(r, 4), round(p, 5)]
    return corr_df.reindex(corr_df.r.abs().sort_values(ascending=False).index)
```

```python
corr_df = bivariate_stat(df, "SalePrice")
corr_df
```
